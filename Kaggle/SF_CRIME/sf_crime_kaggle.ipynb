```
import pandas as pd
train = pd.read_csv("/content/dataset/train.csv")
test = pd.read_csv("/content/dataset/test.csv")
```

```
train.head()
```

```
test.head()
```

```
train.info()
```

```
test.info()
```

```
train.drop(columns = ['Resolution', 'Address', 'Descript'], inplace= True, axis= 1)
test.drop(columns= ['Address'], inplace= True, axis= 1)
```

```
train.head()
```

```
train['DayOfWeek'].replace(to_replace= ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
                           value= [i for i in range(0,7)], inplace= True)
test['DayOfWeek'].replace(to_replace= ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
                           value= [i for i in range(0,7)], inplace= True)
```

```
train.head()
```

```
train_dummies = pd.get_dummies(train['PdDistrict'])
test_dummies = pd.get_dummies(test['PdDistrict'])

train = pd.concat([train, train_dummies], axis= 1)
train.drop(columns= ['PdDistrict'], inplace= True, axis= 1)

test = pd.concat([test, test_dummies], axis= 1)
test.drop(columns= ['PdDistrict'], inplace= True, axis= 1)
```

```
train.head()
```

```

df = pd.DataFrame(columns= ['Year', 'Month', 'Day', 'Hour', 'Minute'])

l1 = list(train['Dates'])
for i in range(len(l1)):
    l1[i] = l1[i][:4:]
df['Year']=l1

l1 = list(train['Dates'])
for i in range(len(l1)):
    l1[i] = l1[i][5:7:]
df['Month']=l1

l1 = list(train['Dates'])
for i in range(len(l1)):
    l1[i] = l1[i][8:10:]
df['Day']=l1

l1 = list(train['Dates'])
for i in range(len(l1)):
    l1[i] = l1[i][11:-6:]
df['Hour']=l1

l1 = list(train['Dates'])
for i in range(len(l1)):
    l1[i] = l1[i][-5:-3:]
df['Minute']=l1

train = pd.concat([train, df], axis= 1)
train.drop(columns=['Dates'],inplace=True, axis=1)
```

```
df = pd.DataFrame(columns= ['Year', 'Month', 'Day', 'Hour', 'Minute'])

l1 = list(test['Dates'])
for i in range(len(l1)):
    l1[i] = l1[i][:4:]
df['Year']=l1

l1 = list(test['Dates'])
for i in range(len(l1)):
    l1[i] = l1[i][5:7:]
df['Month']=l1

l1 = list(test['Dates'])
for i in range(len(l1)):
    l1[i] = l1[i][8:10:]
df['Day']=l1

l1 = list(test['Dates'])
for i in range(len(l1)):
    l1[i] = l1[i][11:-6:]
df['Hour']=l1

l1 = list(test['Dates'])
for i in range(len(l1)):
    l1[i] = l1[i][-5:-3:]
df['Minute']=l1

test = pd.concat([test, df], axis= 1)
test.drop(columns=['Dates'], inplace= True, axis=1)
```

```
X = train.drop(columns=['Category'], axis=1)
y = pd.get_dummies(train['Category'])
```

```
X.head()
```

```
y.head()
```

```
X = X.astype(float)
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1,random_state=42)
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```
from tensorflow import keras
from keras.layers import Dense,Activation
from keras.models import Sequential

model = Sequential()
model.add(Dense(100, input_shape=(X.shape[1],)))
model.add(Activation('relu'))
model.add(Dense(100))
model.add(Activation('relu'))
model.add(Dense(100))
model.add(Activation('relu'))
model.add(Dense(80))
model.add(Activation('relu'))
model.add(Dense(60))
model.add(Activation('relu'))
model.add(Dense(39))
model.add(Activation('softmax'))
model.summary()
```

```
model.compile(optimizer= 'adam',
              loss = 'categorical_crossentropy',
              metrics=['accuracy'])
```

```
import os
if(os.path.exists('/content/trained_model.h5')):
  # load the model architecture and weights
  model.load_weights("/content/trained_model.h5")
else:
  # Start to train model
  r = model.fit(X_train,y_train, 
        batch_size = 32, 
        epochs = 20, 
        verbose = 2, 
        validation_data=(X_test,y_test))
  # save the model with h5py
  import h5py
  model.save('/content/trained_model.h5')

```

```
test_x = test.drop(['Id'], axis = 1)
test_x = test_x.astype(float)
```

```
pred = model.predict(test_x)
```

```
import numpy as np
s = np.max(pred, axis =1).reshape(-1,1)
final = np.array((pred == s), dtype = 'int32')
result = pd.DataFrame()
result['Id'] = test['Id']
sample=pd.read_csv('/content/dataset/sampleSubmission.csv')
column_names = list(sample.columns)
column_names.remove('Id')
for i, e in enumerate(column_names):
    result[e] = final[:,i]
result.head()
```
